```python
import numpy as np

data = np.array([2, 3, 7, 8, 10])
n = len(data)

m1_prime = np.mean(data**1)
m2_prime = np.mean(data**2)
m3_prime = np.mean(data**3)
m4_prime = np.mean(data**4)

print("First moment (m1'):", m1_prime)
print("Second moment (m2'):", m2_prime)
print("Third moment (m3'):", m3_prime)
print("Fourth moment (m4'):", m4_prime)
```

```python
mean = np.mean(data)

m1 = np.mean((data - mean)**1)
m2 = np.mean((data - mean)**2)
m3 = np.mean((data - mean)**3)
m4 = np.mean((data - mean)**4)

print("First central moment (m1):", m1)
print("Second central moment (m2):", m2)
print("Third central moment (m3):", m3)
print("Fourth central moment (m4):", m4)
```

```python
m4_formula = m4_prime - 4*m1_prime*m3_prime + 6*(m1_prime**2)*m2_prime - 3*(m1_prime**4)

print("m4 (computed from formula):", m4_formula)
print("m4 (from central moment):", m4)
```
